# Current analysis with the RBR CTD

Conductivity, temperature, and pressure (depth) form the basic toolkit of oceanographic measurements. Many common CTDs feature an electrode-based conductivity cell, which requires a pump mechanism to actively circulate water past the sensors. This generally translates to overall higher power consumption, as well as noise and vibrations from the moving mechanical parts.

## Load the Data



Data from a CSV file named "sample_CTD_data.csv" is being read into a pandas DataFrame, and custom column names are being specified for the DataFrame. The desired column names have been defined in the "colnames" list. 

The CSV file is being read with specific parameters, including character encoding, and the first 11 rows, often containing header. 

Any rows containing missing values are being removed to clean the data, ensuring that only complete rows are retained. 

The "timestamp" column is being converted to datetime object.

```python

rsk = pyrsktools.RSK('data/sample_CTD_data.rsk')
# Open the RSK file. Metadata is read here
with rsk as rsk:
    # Print a list of all the channels in the RSK file
    #rsk.printchannels()
    # Read data
    rsk.readdata()
    # Derive sea pressure from total pressure
    rsk.deriveseapressure()
    rsk.derivedepth()
    rsk.derivevelocity()
    #Make csv
    #rsk.printchannels()
    rsk.RSK2CSV()
    df = pd.DataFrame(rsk.printchannels())
    #print(rsk.logs)
    #print('')
df.head()
```

```python
colnames=['timestamp','conductivity','temperature','pressure(dbar)',
          'temperature1','dissolved_o2_saturation', 'backscatter',
          'backscatter1','chlorophyll','sea_pressure(dbar)','depth(m)','velocity(m/s)'] 

data = pd.read_csv("sample_CTD_data.csv", encoding="ISO-8859-1", skiprows=11, header=None, names=colnames)
data = data.dropna(axis=0, how="any")
data = data[1:] # drop the first measurrement


#convert the conductivity column to numeric value
cols = data.columns
data[cols[1]] = pd.to_numeric(data[cols[1]])

#make timestamp the index of the datset 
data['timestamp'] = pd.to_datetime(data['timestamp'])
#data.set_index("timestamp", inplace=True)
```

## Reshape data for spectral analyses

Define parameters such as the sample frequency (Hz), burst length (rl), and time intervals. 

Reshape the data for burst analysis.

```python
df_select = data

Hz = 8. # sample freq.
rl = 2 # minutes
dt = 1./Hz
bl = int((rl/dt)*60.) # burst length
nbl = int(math.floor(len(df_select)/bl))

# Reshape data
time = df_select.timestamp[0:nbl*bl].values.reshape((bl,nbl), order='F')
press = df_select['pressure(dbar)'][0:nbl*bl].values.reshape((bl,nbl), order='F')
depth = df_select['depth(m)'][0:nbl*bl].values.reshape((bl,nbl), order='F').mean(axis=0)

mean_pressure = press.mean(axis=0)
mean_pressure = mean_pressure * 0.0980665
```

### Step 2: Use the plot_waves function to plot wave components
#### 1. Mean Water Level Plot

    Use plt.subplots() to create a single-row, single-subplot figure (ax0).
    Plot the mean water level data against time using ax0.plot().
    Adjust tick parameters and set the y-axis label as "Mean water level (m)".

#### 2.  Wave Height Plots

    Create a new figure with two subplots (ax0 and ax1) to visualize wave height metrics.
    Plot Hm0, Hrms, and Hmax on ax0 with custom line styles and labels.
    Set the y-axis label as "Wave height (m)" and add a legend.
    Plot Tm02 and Tm01 on ax1 with similar customization.
    Set the y-axis label as "Wave period (s)" and add a legend.
    Specify the x-axis label and set the x-axis limits based on your data range.

```python
def plot_waves(data, name, band = 'wind', save = False):
    #set up variables:
    if band == 'wind':
        wavetype = 'W'
    if band == 'grav':
        wavetype = 'G'
    if band == 'IG':
        wavetype = 'IG'
    else:
        print('band needs to be wind, grav or IG')
    
    
    
    time = data['time'].values
    mean_water_level = data['mean_water_level(m)'].values

    fig, (ax0, ax1, ax2) = plt.subplots(figsize=(9,6), nrows=3, sharex=True, 
                                   gridspec_kw={'hspace': 0.})
    ax0.set_title(f'{name} {band} waves')
    ax0.plot(time, mean_water_level, lw=2, c='red')
    ax0.tick_params(axis='both', which='major', labelsize=6)
    ax0.set_ylabel('Mean water level (m)', fontsize=8)   
    plt.xlabel('Time', fontsize=8)
    plt.xlim(time[0],time[-1])
    ax0.grid()
    plt.rc('legend',**{'fontsize':7})
    
    
    fig.set_figheight(10) 
    # Change figsize for ax0 to make it smaller
    ax0.set_position([ax0.get_position().x0, ax0.get_position().y0, ax0.get_position().width, 0.1])
    
    ax1.plot(time, data[f'{wavetype}_Hm0'], lw=1, c='k', label='Hm0')
    ax1.plot(time, data[f'{wavetype}_Hrms'], lw=1, c='k', ls='-.', label='Hrms')                  
    ax1.plot(time, data[f'{wavetype}_Hmax'], lw=1, c='k', ls=':', label='Hmax')
    

    ax1.tick_params(axis='both', which='major', labelsize=6)
    ax1.set_ylabel('Wave height (m)', fontsize=8)
    plt.xlabel('Time', fontsize=8)
    ax1.legend(loc=0,frameon=False)
    ax1.grid()
        
    ax2.plot(time, data[f'{wavetype}_Tm02'], lw=1, c='k', label='Tm02')
    ax2.plot(time, data[f'{wavetype}_Tm01'], lw=1, c='k', ls='-.', label='Tm01')
    ax2.tick_params(axis='both', which='major', labelsize=6)
    ax2.set_ylabel(f'{band} Wave period (s)', fontsize=8)
    plt.xlabel('Date', fontsize=8)
    plt.xlim(time[0],time[-1])
    ax2.legend(loc=0,frameon=False)
    plt.grid()
    if save:
        plt.savefig(f'{name}_{band}_waves.png')
    
    plt.show()
    
```

### Plot Swell waves:

```python

plot_waves(df, 'PT', band = 'grav', save = True)
```
