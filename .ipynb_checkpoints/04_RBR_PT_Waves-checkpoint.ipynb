## Reshape data for spectral analyses

Define parameters such as the sample frequency (Hz), burst length (rl), and time intervals. 

Reshape the data for burst analysis.

```python
Hz = 8. # sample freq.
rl = 15 # minutes
dt = 1./Hz
bl = int((rl/dt)*60.) # burst length
nbl = int(math.floor(len(df_select)/bl))

# Reshape data
time = df_select.timestamp[0:nbl*bl].values.reshape((bl,nbl), order='F')
press = df_select['pressure(dbar)'][0:nbl*bl].values.reshape((bl,nbl), order='F')
depth = df_select['depth(m)'][0:nbl*bl].values.reshape((bl,nbl), order='F').mean(axis=0)

mean_pressure = press.mean(axis=0)
mean_pressure = mean_pressure * 0.0980665
```

```python
def plot_waves(data, name, band = 'wind', save = False):
    #set up variables:
    if band == 'wind':
        wavetype = 'W'
    if band == 'grav':
        wavetype = 'G'
    if band == 'IG':
        wavetype = 'IG'
    else:
        print('band needs to be wind, grav or IG')
    
    
    
    time = data['time'].values
    mean_water_level = data['mean_water_level(m)'].values

    fig, (ax0, ax1, ax2) = plt.subplots(figsize=(9,6), nrows=3, sharex=True, 
                                   gridspec_kw={'hspace': 0.})
    ax0.set_title(f'{name} {band} waves')
    ax0.plot(time, mean_water_level, lw=2, c='red')
    ax0.tick_params(axis='both', which='major', labelsize=6)
    ax0.set_ylabel('Mean water level (m)', fontsize=8)   
    plt.xlabel('Time', fontsize=8)
    plt.xlim(time[0],time[-1])
    ax0.grid()
    plt.rc('legend',**{'fontsize':7})
    
    
    fig.set_figheight(10) 
    # Change figsize for ax0 to make it smaller
    ax0.set_position([ax0.get_position().x0, ax0.get_position().y0, ax0.get_position().width, 0.1])
    
    ax1.plot(time, data[f'{wavetype}_Hm0'], lw=1, c='k', label='Hm0')
    ax1.plot(time, data[f'{wavetype}_Hrms'], lw=1, c='k', ls='-.', label='Hrms')                  
    ax1.plot(time, data[f'{wavetype}_Hmax'], lw=1, c='k', ls=':', label='Hmax')
    

    ax1.tick_params(axis='both', which='major', labelsize=6)
    ax1.set_ylabel('Wave height (m)', fontsize=8)
    plt.xlabel('Time', fontsize=8)
    ax1.legend(loc=0,frameon=False)
    ax1.grid()
        
    ax2.plot(time, data[f'{wavetype}_Tm02'], lw=1, c='k', label='Tm02')
    ax2.plot(time, data[f'{wavetype}_Tm01'], lw=1, c='k', ls='-.', label='Tm01')
    ax2.tick_params(axis='both', which='major', labelsize=6)
    ax2.set_ylabel(f'{band} Wave period (s)', fontsize=8)
    plt.xlabel('Date', fontsize=8)
    plt.xlim(time[0],time[-1])
    ax2.legend(loc=0,frameon=False)
    plt.grid()
    if save:
        plt.savefig(f'{name}_{band}_waves.png')
    
    plt.show()
    
```

## Calulate Energy Density for each wave type

Use calculate_area_in_range function  to calculate the area under a curve within a specified frequency range using the trapezoidal rule. 

frequency and energy_density are arrays representing the frequency values and corresponding energy density values, respectively.

The lower_freq and upper_freq parameters specify the lower and upper bounds of the frequency range for which you want to calculate the area under the curve.

np.where((frequency >= lower_freq) & (frequency <= upper_freq))[0] is used to find the indices of data points that fall within the specified frequency range. This line of code creates a boolean mask using the conditions (frequency >= lower_freq) and (frequency <= upper_freq). The mask is then used to index the array and retrieve the indices of the data points that meet the condition. The [0] at the end extracts the indices from the tuple returned by np.where.

Using the np.trapz function, the function calculates the area under the curve defined by subset_energy_density with respect to subset_frequency. The trapezoidal rule is used for this calculation.

```python
def calculate_area_in_range(frequency, energy_density, lower_freq, upper_freq):
    # Find indices of data points within the specified frequency range
    indices = np.where((frequency >= lower_freq) & (frequency <= upper_freq))[0]
    
    # Subset data within the specified frequency range
    subset_frequency = frequency[indices]
    subset_energy_density = energy_density[indices]
    
    # Calculate the area under the curve using the trapezoidal rule for the subset
    area = np.trapz(subset_energy_density, x=subset_frequency)
    
    return area

# Specify the lower and upper frequencies for the range
t_fr_lo = 0.0033
t_fr_hi = 1

# 1 to 3 s
W_fr_lo = 0.33
W_fr_hi = 1

# 3 to 20 s
G_fr_lo = 0.05
G_fr_hi = 0.33

# 20 to 300 s
IG_fr_lo = 0.0033
IG_fr_hi = 0.05

# Calculate the area within the specified frequency range
area_tot = calculate_area_in_range(G_fspec, G_s, t_fr_lo, t_fr_hi)
area_G = calculate_area_in_range(G_fspec, G_s, G_fr_lo, G_fr_hi)
area_IG = calculate_area_in_range(G_fspec, G_s, IG_fr_lo, IG_fr_hi)
area_W = calculate_area_in_range(G_fspec, G_s, W_fr_lo, W_fr_hi)
```

##  Analyzing and visualizing energy density data in the frequency domain

We reload the data and make select the time interval that we want to do our spectral analysis for.

We calculate the energy density for those for a pre-defined timedelta (30 minute burst intervals) between our start and end date.

Finally we plot the energy densities through time.

```python


start_time = pd.to_datetime('2022-10-06 23:00:00')
end_time = pd.to_datetime('2022-10-07 11:00:00')


Hz = 2. # sample freq.
rl = 30 # minutes
dt = 1./Hz
bl = int((rl/dt)*60.) # burst length


s_G_s_list= []
s_G_fspec_list = []


timedelta = '30T'
  
start_times = pd.date_range(start= pd.Timestamp(start_time).round(timedelta), end = pd.Timestamp(end_time).round(timedelta), freq=timedelta)

end_times = start_times + pd.Timedelta(timedelta)

start_times_list = start_times.strftime('%Y-%m-%d %H:%M:%S').tolist()

end_times_list = end_times.strftime('%Y-%m-%d %H:%M:%S').tolist()


for i in range(len(start_times_list[:-1])): 

    start = pd.to_datetime(start_times_list[i])
    end = pd.to_datetime(end_times_list[i])
    
    #print(f'Start Time: {start}. End Time: {end}. Duration of Analysis {end-start}')
    
    s_df_select = data[(data['timestamp']>=start)&(data['timestamp']<=end)].reset_index()

    #print(s_df_select.head(1))
    nbl = int(math.floor(len(s_df_select)/bl))

   

    # Reshape data
    #time = df_select.timestamp[0:nbl*bl].values.reshape((bl,nbl), order='F')
    press = s_df_select['pressure(dbar)'][0:nbl*bl].values.reshape((bl,nbl), order='F')

    #print("pressusre = " , press)
    s_Gwave_s = []
    #s_Gwave_time = []
    s_Gwave_fspec = []

    for k in range(nbl):

        p = PT_power(press[:,k],dt,1) 
        #s_Gwave_time.append(time[0,k])
        s_Gwave_s.append(p[0])
        s_Gwave_fspec.append(p[1])


    s_G_s = np.atleast_2d(np.asarray(s_Gwave_s)).T.conj()[:,0].real
    #s_G_time = np.asarray(Gwave_time)
    s_G_fspec = np.atleast_2d(np.asarray(s_Gwave_fspec)).T.conj()[:,0].real


    s_G_s_list.append(s_G_s)
    s_G_fspec_list.append(s_G_fspec)
    
fig, ax0 = plt.subplots(figsize=(10,8), nrows=1, sharex=True, 
                        gridspec_kw={'hspace': 0})
ax0.plot(s_G_fspec_list, s_G_s_list, lw=1)
ax0.tick_params(axis='both', which='major', labelsize=10)
ax0.set_ylabel('Energy density (m^2 Hz^-1)', fontsize=12)
plt.xlabel('Frequency (Hz)', fontsize=12)
plt.xlim(G_fspec.min(),s_G_fspec.max())
plt.xlim(0,0.2)
plt.ylim(0,1.5)
ax0.legend(loc=1,frameon=False)

plt.tight_layout()

plt.show()
```
